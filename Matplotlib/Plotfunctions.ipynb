```python
#plt.plot(x, y, color='color name', lifestyle='line_style', linewidth=value, marker='marker symbol', label='label name')

import matplotlib.pyplot as plt

months = [1,2,3,4,5]
sales = [12000, 10000, 20000, 15000, 24500]

plt.plot(months, sales, color = 'red', linestyle='--', linewidth= 2, marker = 'o', label='2025 sales data')

plt.title("Monthly sales Data report")
plt.xlabel("Number of months")
plt.ylabel("Sales per month ")
plt.legend(loc="lower right", fontsize=10)
plt.grid(color="grey", linestyle=':', linewidth=1)
plt.xlim(0,6)
plt.ylim(500, 30000)
plt.xticks([1,2,3,4,5], ['Jan','Feb','Mar','Apr','May'])
plt.show()
```
