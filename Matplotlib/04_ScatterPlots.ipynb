```python
#plt.scatter(x,y, color='color_name', marker='marker style', label='label name')

import matplotlib.pyplot as plt

hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50,56,63,69,72,81,86,95]

#scatter plot
plt.scatter(hours_studied, exam_scores, color='red', marker='o', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relatonship Between Exam time & Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
