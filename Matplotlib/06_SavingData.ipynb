```python
#savefig('filename.extension', dpi = value, bbox_inches = 'tight' )

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [43,32,64,24,12,47]

# create plot

plt.plot(x,y,color='green', marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.savefig('Line_Plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
