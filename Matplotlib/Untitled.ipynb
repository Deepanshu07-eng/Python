```python
#plt.subplot(nrows, nvols, index)

import matplotlib.pyplot as plt

x = [1,2,3,4]
y= [23, 43, 53, 34]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt

#fig, ax = plt.subplot(nrows, ncols, figsize=(width, height))
fig, ax = plt.subplots(1,2,figsize=(10,5))

x = [1,2,3,4]
y= [23, 43, 53, 34]

ax[0].plot(x,y)
ax[0].set_title('Line Chart')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

plt.tight_layout() #always use before plt.show() function
plt.show()

```
