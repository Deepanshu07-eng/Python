```python
import matplotlib.pyplot as plt
x = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri' ]  #x-axis
y = [10, 15, 2, 24, 18 ]  #y-axis

plt.plot(x,y)
plt.title("Bakery Sales This Week")
plt.xlabel("Day of the Week")
plt.ylabel("Sales per day")
         
plt.show
```
